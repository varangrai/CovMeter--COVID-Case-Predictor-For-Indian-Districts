```
def cleaner(liste,df,colname):
  liste=set(liste)
  bl=[]
  for i,el in enumerate(df[colname]):
    if el not in liste: bl.append(i)
  df=df.iloc[bl,:]
  df.reset_index(drop=True,inplace=True)
  df.head()
  return df
list_state=['Assam (3 Migrated)','Haryana (Include 14 Italians and Foreign Returnee in Unknown)',
"Himachal Pradesh (11 Migrated)","Madhya Pradesh* does not include migrant's data",
'Maharashtra* does not include 25 migrant workers data (23 active)','District Wise Total (w/out unknown)',
'Unknown Total','District Wise Total','MOHFW','Check']
list_dist=['Andhra Pradesh','Unknown (migrants and foreign returnees)','Arunachal Pradesh','Unknown',
       'Assam','Bihar','Goa','Gujarat','Others']
active=cleaner(list_state,active,'State')
active=cleaner(list_dist,active,'District')
# active=pd.concat([active,activeback])
```

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(total_cleaned['State'])
final_array_state=np.zeros((5,357,736)) 
np.array(total_cleaned.iloc[0,2:])
for i in range(736):
  ar1=np.array(total_cleaned.iloc[i,2:])
  ar2=np.array(deaths_cleaned.iloc[i,2:])
  ar3=np.array(recoveries_cleaned.iloc[i,2:])
  ar4=np.array(active_cleaned.iloc[i,2:])
  ar5=le.transform([total_cleaned['State'][i]])[0]*np.ones((1,357))
  final_array_state[0,:,i]=ar1
  final_array_state[1,:,i]=ar2
  final_array_state[2,:,i]=ar3
  final_array_state[3,:,i]=ar4
  final_array_state[4,:,i]=ar5
np.save('3dstate',final_array_state)
```

```
final_array_state_2d=np.reshape(final_array_state,(5,357*736))
np.save('3dstate',final_array_state)
np.save('2dstate',final_array_state_2d)
```
